```python
# Indexing in numpy Arrays

# import numpy as np
# list=[1,2,3,4]
# array1=np.array(list)
# print(array1[0]) // 1
# print(array1[-1]) // 4

# for 2D array
import numpy as np
list=[[1,2,3],[4,5,6],[7,8,9]]
array1=np.array(list)
print(array1[1,2)
print(array1[0,:)
print(array1[:,1)
```

```python
# Arithmetic Operations on numpy arrays
# 1) Addition(+) 2) Subtraction(-) 3) Multiplication(simple multiplication(*) karega)
# 4) metrics multiplication(@) 5) Division(/) 6) floor division(//) 7)Exponentiations(**)
# 8) Modulo(% Remainder deta hai) 9) Transpose

import numpy as np
x=np.array([[1,2,3],[4,3,2]])
y=np.array([[4,6,7],[2,3,4]])
# z=x+y
print(x.transpose())
```

```python
# Sorting in numpy arrays
1) np.sort() // it will returns a sorted copy of an array
2) np.argsort() // will return the indices that would sort an array
```

```python
import numpy as np
x=np.array([4,5,3,2,7])
y=np.sort(x) #// ascending oredr me sort karega
# y=np.sort(x)[::-1] // decending order me sort hoga
print(y)
```

```python
import numpy as np
x=np.array([4,5,3,2,7])
y=np.argsort(x) # sorted array ka index return karega
print(y)
```
